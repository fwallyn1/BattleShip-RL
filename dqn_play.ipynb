```python
from dqn_agent import DQNAgent
from battleship_env import BattleshipEnv
env = BattleshipEnv()
agent = DQNAgent(env)
agent.epsilon=0
agent.load("battleship-dqn.h5")

# play one game
state = env.reset()
env.render()
done = False
hit = 0
while not done:
    action = int(agent.act(state))
    next_state, reward, done, info = env.step(action)
    env.render()
    state = next_state
    hit += 1
env.render()
print("Number of hits to win the game:", hit)
```
